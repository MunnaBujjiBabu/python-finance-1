## 1.1 Population and Sample

All models try to get a grasp about the population but large majority of them does this using samples. As population is a set of events, which is of interest for modeling or for other purposes such as pool, it requires considerable amount of labor to collect the data for every events. To be more specific, to understand the political tendency of a city, it would be quite though task to talk to everyone and ask their opinion who have voting right.

Thus, instead, we can get a random subset from the population. We call this subset a `sample`. Sampling makes us to access information about the entire population and we take advantage of this information for inferences about the population. Please note that even though we use sampling, we are trying to generalize it and treat the result we get from sample as the one from population.

## 3. Central Tendencies Measures

As we are dealing with a bunch of data instead of the data points, it is normal to try to understand the characteristics of this bunch of data. Measures of central tendency is a tool to accomplish that. It is a summary statistics by which we get a certain sense about the distribution of the data.

The central tendency measures are:
* Mean (or average), 
* Median, 
* Mode

### 3.1. Mean

`Mean` is basically summing up every values of a variable and dividing it by the total number of observation. It is also known as expected value:

$$ mean(x)=sum(x)/n $$

There are two ways to calculate it via Python: **Custom-based** and **Built-in** function. Let's create a data frame and fill it with some values of our choice. Please note that before doing that we need to introduce an empty dataframe:

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame()
df['random'] = [35, 11, 22, 25,25, 10, 16, 87, 45]
```

```python
# custom-based
print(sum(df['random']) / len(df['random']))

# Built-in
print(np.mean(df['random']))
```

It is easy to understand the mean concept. However, it is worth noting that this measure is sensitive to outliers, that is to say, if our sample includes many excessive values, it inevitably affects the mean of this group. And also, please keep in mind that every measures comes with a cost. So, to decide which one to use largely depends on the data you have. In other words, because the structure of a data can be best represented using the mean of central tendency measures, while another data structure can be best represented by `mode` or `median`.

### 3.2. Median

`Median` is another central tendency measures. When the data is in a descending order, it is the value right in the middle if we have odd number of observation. If the number of observation is even, median is the average of the two values in the middle.

Here, we will see two different methods to calculate the median. In the first one, we use `statistics` package and in the second one, we refer to numpy.

```python
import statistics

print(statistics.median(df['random']))

print(np.median(df['random']))
```

### 3.3. Mode

The `mode` is the value that occurs most frequently in the dataset. In other words, the mode analysis enable us to find if there is more than one cluster in the dataset. If two or more values that repeats most, then it is called `double mode`.

Let's use the `statistics` package first to calculate mode.

```python
import statistics
statistics.mode(df['random'])
```

Here is how you can calculate mode using numpy. First, we identify the number of counts of every single values in the dataset. Then, simply by applying `argmax`, we find index and finally we apply this index to values to find the most frequently occured value.

```python
(values, counts) = np.unique(df['random'], return_counts=True)
index = np.argmax(counts)
values[index]
```

## 4. Dispersion Measures

It is not possible to fully understand the distribution by having the central tendency measures. Therefore, we need more information about the distribution. Thus far, we only have information about the middle value of the data set but how about the spreadness of it. `Dispersion` does serve this purpose. It informs us the extent to which the distribution is dispersed or stretched.

Three dispersion measures are:

* Variance
* Standard Error
* Standard Deviation

### 4.1. Variance

The variance of a variable defines how each value differ from the central tendency and how each value differs from another. If all the values in a variable are close to the central tendency, the variance would be very small. If the values are far from the central tendency, the variance would be very high.

The variance is calculated by the sum of the squares of the difference of each data point to the average divided by one minus the number of observations.

$$ v=\frac{\sum (x-\mu)^ 2}{n-1}$$

where $x$ is data, $\mu$ represents the mean of the data and n is the number of observations.

Maybe, there are 2 points that arouse curiosity: 
    * First, why we take the square of the term in the numerator?
    * Second, why we divide it by n-1 instead of n?

In response to the first question, the sum would be zero if only the difference between data and the mean is calculated because half of the data would be negative and the other half would be positive. By squaring the data, we make the negative values positive. 

The answer of the second question lies in the bias correction. Using `n` in the denominator would lead to a bias in estimating the population variance from a sample. So, multiplying the uncorrected sample variance by the factor **n/n-1** gives an unbiased estimator of the population variance. 

```python
print(np.var(df['random'], ddof=1))
print(df['random'].var())
```

### 4.2. Standard Deviation

One of the most used indicators to measure volatility is the square root of the variance, we call it `standard deviation`. There are some mathematical benefits of taking the square root of the variance, but we will see this in more detail when describing the `central limit theorem` part. The formula of standard deviation:

$$s=\sqrt v$$

Numpy provides us a built-in function to calculate the standard deviation, which is `np.std()`. Numpy has a deceptive nuance when calculating the standard deviation of the population. Instead of dividing by `n-1`, it divides by `n`. To calculate the standard deviation of the population, we manually introduce the `ddof` (degrees of freedom) parameter:

```python
np.std(df['random'], ddof=1)
```

### 4.3 Standard Error

It is very common to confuse the `standard error` with standard deviation.  The former is used to calculate the dispersion of a sample. However, the latter gives us a sense of dispersen of the population.

For instance, we use standard error in surveys.  It is also commonly known as error margin. The formula is as follows:

$$
se=\frac{s}{\sqrt n}
$$

```python
np.std(df['random'] ,ddof=1) / np.sqrt(len(df['random']))
```

Now, let's observe what it looks like to have higher or lower variance dataset. In order to have these datasets, we generate my own datasets that we will dig deeper in the upcoming part. As always, we first create an empty data frame and then generate data from normal distribution using built-in numpy function:

For the sake of practice, our sample datasets has the same mean, which is 3 and the standard deviation is 10 and 1 in the high and low variance dataset, respectively.

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
sns.set()
pop=pd.DataFrame()

pop['high']=np.random.normal(3, 10, 10000)
pop['low']=np.random.normal(3, 1, 10000)
pop.hist()
plt.show()
```

## 5. Central Limit Theorem and Law of Large Number

In this part of the course, we will discuss two important theorem about statistics: `Central Limit Theorem` and `Law of Large Numbers`.

### 5.1. Central Limit Theorem

As we briefly touch before, it is better to know the distribution of a data before modeling it and, most of the time, data is assumed to have normally distributed. This is because it is easy to model normally distributed data. Thus, many researchers struggle to have a normally distributed data even if they do not have one. At this point, Central Limit Theorem allows us to model non-normally distributed data.

The central limit theorem states that the sampling distribution of the mean of any independent, random variable will be normal or nearly normal, if the sample size is large enough. So, central limit theorem highlights the importance of normal distribution.

The only condition needed to be met is that samples should be independent and identically distributed (iid).

```python
from numpy.random import randint
plt.figure(figsize=(10, 5))


sns.set()
means1=[np.mean(randint(1, 7, 60)) for i in range(100)]
means2=[np.mean(randint(1, 7, 60)) for i in range(10000)]
means3=[np.mean(randint(1, 7, 60)) for i in range(1000000)]

plt.subplot(1,3,1)
plt.hist(means1)
plt.title('Low Number of Samples')
plt.subplot(1,3,2)
plt.hist(means2)
plt.title('Moderate Number of Samples')
plt.subplot(1,3,3)
plt.hist(means3)
plt.title('Large Number of Samples')

plt.show()
```

```python
from scipy import stats

print(stats.describe(means1),'\n')
print(stats.describe(means2),'\n')
print(stats.describe(means3),'\n')
```

Great!, as we incrase the sample size, we approach more and more to the population mean. Because, `means1`has a mean of 3.4915, `means2` has a mean of 3.4965, and finally `means3` has a mean of 3.500 85, which is quite close to the population mean.

### 5.2. Law of Large Number

`Law of Large Number` stresses the importance of large sample size. Accordingly, as we increase the number of sample size, the sample mean becomes very close to the population mean.

Now, let's see how the Law of Large Number works:

```python
results1 = []
results2 = []
results3 = []

for trials1,trials2,trials3 in zip(range(1,100),range(1,10000),range(1,1000000)):
    trials_1 = np.random.randint(1, 7, trials1)
    trials_2 = np.random.randint(1, 7, trials2)
    trials_3 = np.random.randint(1, 7, trials3)
    mean_of_trials1 = trials_1.mean()
    mean_of_trials2 = trials_2.mean()
    mean_of_trials3 = trials_3.mean()
    results1.append(mean_of_trials1)
    results2.append(mean_of_trials2)
    results3.append(mean_of_trials3)

df1 = pd.DataFrame({ 'trials1' : results1})
df2 = pd.DataFrame({ 'trials2' : results2})
df3 = pd.DataFrame({ 'trials3' : results3})


df1.plot()
plt.title('Low Sample Size')

df2.plot()
plt.title('Moderate Sample Size')

df3.plot()
plt.title('Large Sample Size')
plt.show()

```

Here, it is readily observable that, in the third plot, the data oscillates around 3.5, which is the population mean. As expected, the volatility is larger in the first and second plots.

## 6. Distributions 

Large number of scientific methods are based on the assumption of `normal distribution`. There are two main reasons of why normal distribution is so popular:

    * It can be easily summarized by only two statistical parameters (mean ($\mu$) and standard deviation ($\sigma$).
    * The area under the curve is 1, which makes it easy to calculate the probability of discrete results within the distribution.
    * In many areas such as psychology and finance, data are closely distributed to normal.
    * In general, any variable that measures the outcome of many minor effects is independently close to the normal distribution.
    * Many of the common indicators (z-score) and statistical tests (t-tests, ANOVAs, bell curve) are based on the assumption of normal distribution of data. Thus, the normal distribution is easy to use and is therefore widely used.



In this course we will study normal distribution along with some other prominent distributions, which are listed below:

* Normal Distribution
* Bernoulli Distribution
* Binomial Distribution
* Exponential Distribution
* Poisson Distribution

### 6.1. Normal Distribution

Normal distribution, also known as `Gaussian` distribution following the discovery by the German mathematician `Carl Friedrich Gauss` describes many cases that we encounter in daily life. Normal distribution can be characterized by bell shape, i.e., values are symmetrically clustered around the mean. The probability density function of normal distribution is:

$$
f(x|\mu,\sigma^2)=\frac{1}{\sqrt{2\sigma^2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

where $\mu$ and $\sigma$ represent the mean and standard deviation,respectively. It is important to note that these two parameter characterize the normal distribution. In other words, if these parameters are a priori known, the information in a normal distribution can be summarized.

The question of what makes normal distribution so important can be addressed by the following facts:

* Two parameters (mean ($\mu$) and standard deviation ($\sigma$) are enough to summarize it.
* It is widely used in many areas ranging from biology to finance in which many data follows normal distribution.
* Many statistical approaches such as `z-score`, `t-tests`, and `ANOVA` are based on the assumption of normal distribution of data.

In this part of the course, we also learn how to generate random data from a specific distribution. To do that, we use numpy. Before proceeding, let us describe an important concept in random number generating, which is called `seed`. 

In generating a random sample, we need a starting point, known as  seed value, to be able to have a consistent samples. The seed value can be chosen arbitrarily. In the following example in which we draw samples from normal distribution, we select `1` as seed.

```python
mean = 0 
sd = 1
n = 100000
np.random.seed(1)
df = pd.DataFrame({'Normal Distribution': np.random.normal(mean, sd, n)})

df.hist()
plt.show()
```

As `empirical rule of statistics`suggests that approximately 68% of the values in a normally-distributed variable fall within one standard deviation above or below the mean, 95% of the values fall within two standard deviations of the mean, and 99.7% of the values fall within three standard deviations of the mean.

### 6.2. Bernoulli Distribution

`Bernoulli Distribution` models an event with two possible outcomes such as a coin flip. As the possible outcomes are discrete, the distribution can be understood by probability mass function (pmf). In a nutshell, the probability of each outcome is also between 0 and 1, and the sum of all the probabilities is equal to 1.

The probability mass function can be formulated as:

$$
f(k|p)=\begin{cases}
p&\text{if }k=1\\
1-p&\text{otherwise}\\
\end{cases}$$

where p is probability and k is the probability of outcome.

Here, we again will use random number generating process. However, it may be surprising to use `.binomial()` function to generate Bernoulli distribution. The reason is simple. Binomial distribution is nothing but a repeated and independent Bernoulli experiment over and over again. Therefore, there is no harm to use `.binomial()` as long as we have n=1 trials, which is the first input of `.binomial()` function. The second and third input of `.binomial()` are probability of outcome and output shape, respectively.

```python
bernoulli1= np.random.binomial(1, .5, 10)
bernoulli2= np.random.binomial(1, .5, 100)

plt.subplot(1,2,1)
plt.hist(bernoulli1)
plt.title('After 10 times')
plt.subplot(1,2,2)
plt.hist(bernoulli2)
plt.title('After 100 times')

plt.show()
```

### 6.3. Binomial Distribution


A `binomial distribution` counts the number of successes when an event with two or more distinct possible outcomes is repeated many times. So, it sounds like Bernoulli experiment and you are partially right as we are iterating the independent Bernoulli experiment over and over again.

$$
f(k|n,p)={n\choose k} p^k(1-p)^{n-k}$$

where `n` is total number of events each having p probability. Now to differentiate the Bernoulli trial from the Binomial trial, we just need to revise the first parameter of `binomial()`, which is `n`. To this respect, let's assume that `n` is 20

```python
binomial = np.random.binomial(20, 0.5, 100)

plt.hist(binomial)
plt.title('Binomail Distribution')
plt.show()
```

Let's dig into the process of  Binomial distribution. Suppose that you flip a coin 5 times and it turns out that we have 2 heads and 3 tails. So, the probability of getting head is 0.4 and the probability of getting tail is 0.6. Mathematically speaking,

$$
p(k=2)=C^{5}_2(0.4)^2(0.6)^3$$

### 6.4. Exponential Distribution


Modeling time can be an issue and `Exponential Distribution` is used to model the time. Specifically, it is a probability distribution of time between events.

$$
f(k|p)=\begin{cases}
\lambda e^{-\lambda x}&\text{if }x\geq 0\\
0&\text{otherwise}\\
\end{cases}$$

In order to generate exponentially distributed samples, we use `.random.exponential()` from numpy library. This function has two parameters:
    * Scale: inverse of rate. The default is 1.
    * Size: The shape of the returned array

Inverse of rate corresponds to $\lambda$ that we are going to discuss in Poisson Distribution.

```python
plt.hist(np.random.exponential(scale=3,size=1000))
plt.title('Exponential Distribution')
plt.show()
```

For example, if an event has not occurred after 30 seconds, the conditional probability that occurrence will take at least 10 more seconds is equal to the unconditional probability of observing the event more than 10 seconds after the initial time.

### 6.5. Poisson Distribution

If we want to model an event occuring several times within a specific time interval, `Poisson distribution` is what we are looking for. Suppose that we are willing to understand the performance of bank clerk. It may be a wise way to calculate how many customer he/she is beign taken care of within 1 hour. Number of customers taken care of in one hour can take on values ranging from 0 to very large number. As we are modeling a discrete event, Poisson distribution is characterized by probability mass function:

\begin{equation}
f(k|\lambda)=\frac{\lambda^k e^{-\lambda}}{k!}
\end{equation} 

where $\lambda$ is the rate that events occur during a given time period.

In generating random samples from Poisson distribution, `.random.poisson()` can be used. Two parameters are needed to generate samples:
    * lam (lambda): Number of occurrence within a specific time
    * Size: Output shape

```python
poisson = np.random.poisson(5, 10000)

plt.hist(poisson)
plt.title('Poisson Distribution')
plt.show()
```
