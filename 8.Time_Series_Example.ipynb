## Answer

I will start downloading daily `Facebook` stock close price between 2015-2020. Then, I will visualize and examine the raw data and try to get information.

```python
stock = 'FCB'
start = datetime.datetime(2015,1,1)
end = datetime.datetime(2020,1,1)
fcb_close = yf.download(stock,start=start,end = end, interval='1d')['Close']
```

```python
plt.plot(fcb_close)
plt.title('Facebook Stock Price')
plt.ylabel('$')
plt.xlabel('Date')
plt.show()
```

The line plot given above tells us that there is an upward trend, implying that the mean is not constant over time. Therefore, my first observation is that the `Facebook` closing stock price is not stationary. However, there is more scientific way to test the stationarity, which is called `ADF` test. Let's apply this test and see if it confirms the result we obtain based on the visualization.

```python
stat_test=adfuller(fcb_close)[0:2]
print("The test statistic and p-value of ADF test are {}".format(stat_test))
```

Yes, `ADF` test suggest that this data is non-stationary as I have a p-value of 0.47, larger than 0.05. So, I need to handle the stationarity and the first and foremost method is to apply `first difference`. 

```python
diff_fb_close=fcb_close.diff().dropna()
```

After applying first difference, I draw the line plot to see the differenced price fluctuates around mean. Yes, it does but to make sure I re-run the `ADF` test.

```python
plt.plot(diff_fb_close)
plt.title('Differenced Facebook Stock Price')
plt.ylabel('$')
plt.xlabel('Date')
plt.show()
```

```python
stat_test=adfuller(diff_fb_close)[0:2]
print("The test statistic and p-value of ADF test are {}".format(stat_test))
```

`ADF` test confirms that the differenced `Facebook` stock price data turns out to be a stationary one. Let's move on to the next step!

I arbitrarily pick MA and ARMA to model this data. For MA, I use pacf to define the optimal number of lags and I employ AIC to optimize the ARMA

```python
sm.graphics.tsa.plot_acf(diff_fb_close,lags=30)
plt.show()
```

`PACF` indicates some peaks at 3th, 10th, and 15th lags and it is safe to pick 15th lag as window.

```python
moving_average_fb=diff_fb_close.rolling(window=15).mean()
```

```python
plt.figure(figsize=(10,6))
plt.plot(diff_fb_close.loc[start:end].index, diff_fb_close.loc[start:end], label='Differenced Facebook Stock Price',c='b')
plt.plot(moving_average_fb.loc[start:end].index, moving_average_fb.loc[start:end], label = 'MA',c='r')
plt.title('Predicted Stock Price-Facebook')
plt.xlabel('Date')
plt.ylabel(ylabel='$')
plt.show()
```
